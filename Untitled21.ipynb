```python
# ─────────────────────────────────────────────────────────────────────────────
# CELL 4 — Quantization round-trip + dtype check (K4) — CPU, ~1 s
# Confirms fp16/int8/int4 all dequantize to float16 and storage shrinks.
# ─────────────────────────────────────────────────────────────────────────────
import torch
from kv_quantization import compress_kvcache, decompress_kvcache, cache_size_bytes
layers = [(torch.randn(1,2,7,16,dtype=torch.float16),
           torch.randn(1,2,7,16,dtype=torch.float16)) for _ in range(3)]
print(f"{'prec':5s} {'out_dtype':16s} {'recon_max_err':14s} {'bytes':>7s}")
prev = None
for prec in ['fp16','int8','int4']:
    comp = compress_kvcache(layers, prec)
    deq  = decompress_kvcache(comp, prec)
    dt   = deq[0][0].dtype
    err  = max((deq[i][0].float()-layers[i][0].float()).abs().max().item() for i in range(3))
    sz   = cache_size_bytes(comp, prec)
    assert dt == torch.float16, f"{prec} returned {dt}, expected float16 (K4)"
    if prev is not None: assert sz <= prev, "storage should shrink fp16>=int8>=int4"
    prev = sz
    print(f"{prec:5s} {str(dt):16s} {err:<14.4f} {sz:>7d}")
print("K4 OK: all float16, error fp16<int8<int4, storage shrinks")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# CELL 5 — CORRECTNESS PROOF: cached-decode == full-forward (tiny random model)
# This is the decisive test for K2/K3 (RoPE/cache rewrite). Runs on a 3-layer
# random Qwen2 in seconds. cos must be ~1.0 for the single-cache case.
# ─────────────────────────────────────────────────────────────────────────────
import torch, torch.nn.functional as F
from transformers.models.qwen2.modeling_qwen2 import Qwen2Config
from transformers import DynamicCache
from qwen2 import Qwen2ModifiedForCausalLM

torch.manual_seed(0)
cfg = Qwen2Config(vocab_size=512, hidden_size=128, intermediate_size=256,
    num_hidden_layers=3, num_attention_heads=8, num_key_value_heads=2,
    max_position_embeddings=256, rms_norm_eps=1e-6)
cfg._attn_implementation = "eager"
m = Qwen2ModifiedForCausalLM(cfg).eval()

# (a) Single-cache decode must equal full forward (pure correctness)
doc = torch.randint(0,512,(1,10)); suf = torch.randint(0,512,(1,3))
with torch.no_grad():
    cache = m(doc, use_cache=True).past_key_values
    clen  = cache.get_seq_length()
    lc = m(suf, past_key_values=cache,
           cache_position=torch.arange(clen,clen+3), use_cache=True).logits[0,-1]
    lf = m(torch.cat([doc,suf],1), use_cache=False).logits[0,-1]
cos_single = F.cosine_similarity(lc, lf, 0).item()
print(f"(a) single-cache vs full  cos = {cos_single:.6f}  (must be ~1.000)")
assert cos_single > 0.999, "CORRECTNESS FAIL: cache mechanics broken (K2/K3)"

# (b) Independent-chunk stitching: divergence here is EXPECTED (TurboRAG
#     independent-attention approximation), NOT a bug. Just report it.
def build(ids):
    with torch.no_grad(): return m(ids, use_cache=True).past_key_values
def stack(cs):
    legs=[c.to_legacy_cache() for c in cs]; nl=len(legs[0])
    st=tuple((torch.cat([l[i][0] for l in legs],2),
              torch.cat([l[i][1] for l in legs],2)) for i in range(nl))
    return DynamicCache.from_legacy_cache(st)
h1,h2 = doc[:,:5], doc[:,5:]
st = stack([build(h1), build(h2)]); cl2 = st.get_seq_length()
with torch.no_grad():
    lc2 = m(suf, past_key_values=st,
            cache_position=torch.arange(cl2,cl2+3), use_cache=True).logits[0,-1]
cos_stitch = F.cosine_similarity(lc2, lf, 0).item()
print(f"(b) 2-chunk stitched vs full cos = {cos_stitch:.4f}  (EXPECTED < 1.0 — this")
print("    independent-attention gap is the very effect the paper measures)")
print("\nK2/K3 OK: cache+RoPE mechanics correct; stitching gap is by design")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# CELL 9 — Metrics table + coherence checks (path-agnostic)
# Updated for: paired_n column, avg_io_s, C3 hall >= C1 hall direction check.
# ─────────────────────────────────────────────────────────────────────────────
import glob, os, json, math
import pandas as pd

SCRATCH = os.environ.get('SCRATCH_DIR', '/content/turborag_quant')

patterns = [
    f'{SCRATCH}/results/summary_*.csv',
    '/content/turborag_quant/results/summary_*.csv',
    '/content/kvv/results/summary_*.csv',
]
found = []
for p in patterns:
    found.extend(glob.glob(p))
found = sorted(set(found))

if not found:
    print("Still no summary CSV. Checking log for first error:")
    !grep -n "Traceback\|Error:\|error:\|FAILED" /content/mve_run.log | head -20
    raise SystemExit("No CSV — check the log above.")

csv_path = found[-1]
print(f"Reading: {csv_path}\n")
df = pd.read_csv(csv_path)

# Coerce numeric columns (CSV is written with QUOTE_ALL)
for col in ['k', 'n_examples', 'paired_n', 'n_hall', 'n_total',
            'EM', 'contain_EM', 'F1',
            'hallucination_rate', 'entailment_score',
            'avg_ttft_s', 'avg_io_s', 'refusal_rate', 'avg_kv_bytes']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

want = ['dataset', 'k', 'condition', 'n_examples', 'paired_n',
        'refusal_rate', 'EM', 'contain_EM', 'F1',
        'hallucination_rate', 'entailment_score',
        'avg_ttft_s', 'avg_io_s']
cols = [c for c in want if c in df.columns]
print(df[cols].to_string(index=False))

print("\n--- coherence checks ---")
def val(cond, k, col):
    if col not in df.columns:
        return float('nan')
    rows = df[(df['condition'] == cond) & (df['k'] == k)]
    if rows.empty:
        return float('nan')
    v = rows[col].iloc[0]
    try:
        return float(v)
    except (TypeError, ValueError):
        return float('nan')

all_ok = True
for k in sorted(df['k'].dropna().unique()):
    k = int(k)
    f1_c0   = val('C0', k, 'F1')
    f1_c1   = val('C1', k, 'F1')
    hr_c1   = val('C1', k, 'hallucination_rate')
    hr_c3   = val('C3', k, 'hallucination_rate')
    ref_c0  = val('C0', k, 'refusal_rate')
    pn_c1   = val('C1', k, 'paired_n')
    io_c1   = val('C1', k, 'avg_io_s')
    io_c0   = val('C0', k, 'avg_io_s')

    print(f"\nK={k}:")

    # F1 check
    if not math.isnan(f1_c1):
        ok = max(f1_c0, f1_c1) > 0.05
        print(f"  F1: C0={f1_c0:.3f}  C1={f1_c1:.3f}  "
              f"{'OK' if ok else 'LOW — corpus may not match questions'}")
        if not ok:
            all_ok = False

    # Refusal check
    if not math.isnan(ref_c0):
        ok = ref_c0 < 0.5
        print(f"  refusal C0={ref_c0:.2f}  "
              f"({'OK' if ok else 'HIGH — retrieval not finding relevant context'})")
        if not ok:
            all_ok = False

    # ── KEY CHECK: paired_n > 0 means faithfulness scoring actually ran ──
    if not math.isnan(pn_c1):
        pn = int(pn_c1)
        ok = pn > 0
        print(f"  paired_n (C1)={pn}  "
              f"({'OK — faithfulness denominator valid' if ok else 'ZERO — C0-pairing bug still present or all examples refused. Fix paired_mask in evaluate.py.'})")
        if not ok:
            all_ok = False

    # Hallucination direction check (C3 >= C1 expected)
    if not (math.isnan(hr_c1) or math.isnan(hr_c3)):
        direction_ok = hr_c3 >= hr_c1
        print(f"  hall: C1={hr_c1:.3f}  C3={hr_c3:.3f}  "
              f"{'✓ C3>=C1 expected' if direction_ok else '✗ C3<C1 unexpected — check n is large enough'}")
    else:
        print("  hallucination_rate = N/A (HHEM did not run or paired_n=0)")

    # IO vs TTFT check (C0 should have io≈0, C1-C3 have io>0)
    if not math.isnan(io_c0) and not math.isnan(io_c1):
        print(f"  TTFT: C0.io={io_c0:.3f}s (should be 0)  C1.io={io_c1:.3f}s (disk+dequant)")

print(f"\n{'ALL coherence checks passed ✓' if all_ok else 'SOME checks failed — see WARN lines above'}")
```
